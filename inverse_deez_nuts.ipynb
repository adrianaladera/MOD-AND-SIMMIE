# Inverse Problem for our Jazz
## Forward Problem
### Importing libraries and necessary functions

```julia
using LinearAlgebra
using Printf
using Plots

global const µ = 3.986e14  # m^3/s^2

include("eval_f.jl")
include("trapezoidal.jl")
```

### Helper functions

```julia
function f(x)
    feval(x,0,0)
end
```

```julia
function step(x1, x0, h)
    return x1 - x0 - h/2 * f(x1) - h/2 * f(x0)
end
```

### Testing out trapezoidal function
Simple 2D case, not accounting for out-of-planar orbit and constant velocity vector in the $y$-Cartesian direction

```julia
radius = 7e6  # m
velocity = sqrt(µ/radius)


x0 = [radius, 0, 0, 0, velocity, 0] # simple 2D
t0 = 0
ti = 2*pi*radius^1.5/µ^0.5  # Should result in 1 orbit
num_steps = 1000
xs = trapezoidal(f, t0, ti, x0, num_steps);
```

### Plot the orbit of example satellite given its actual system state vector $x_0$

```julia
println("Final energy / initial energy: ",energy(x0)/energy(xs[num_steps+1,:]))
plot(xs[:,1], xs[:,2], label="actual orbit")
```

# Inverse problem
Given the output values (the orbits), estimate the controller parameters (i.e. estimate its system state vector)
### User guesses orbit, inverse problem updates to minimize error

```julia
print("Estimate the float value of radius in meters (m) of the satellite given its output matrix (i.e. 1,2,...,etc. will be multiplied by 10^6): ")
radius_g = readline()
radius_g =  parse(Float64,radius_g)
```

### Threshold parameters
When the minimization should stop updating once the final answer is within a specified tolerance, and the algorithm updates in steps

```julia
tolerance = 10000 #radius tolerance
update_steps = 5000
err_tol = 6000
```

### Computing with inverse guess, show orbit plots of guess with actual

```julia
velocity_g = sqrt(µ/radius_g)

x0_g = [radius_g, 0, 0, 0, velocity_g, 0]
t0_g = 0
ti_g = 2*pi*radius_g^1.5/µ^0.5  # Should result in 1 orbit
num_steps_g = 1000
xs_g = trapezoidal(f, t0_g, ti_g, x0_g, num_steps_g);

plot()
plot!(xs[:,1], xs[:,2], label="actual")
plot!(xs_g[:,1], xs_g[:,2], label="predicted guess", color="red", linewidth=5)
```

### comparing values of the original predicted satellite orbit with actual orbit values

```julia
err_x = abs.(xs[:,1] - xs_g[:,1])
err_y = abs.(xs[:,2] - xs_g[:,2])
max_errx = maximum(err_x)
max_erry = maximum(err_y)
@printf("Maximum error in x (meters): %f\n", max_errx)
@printf("Maximum error in y (meters): %f\n", max_erry)
plot(err_x, err_y)
```

### Using update loop to get predicted state closer to actual state and plot final orbit

```julia
radius_gl = copy(radius_g);
max_errx_g = copy(max_errx);
max_erry_g = copy(max_erry);
xs_gl = [];
# while (radius_gl > radius + tolerance && max_errx_g > err_tol && max_erry_g > err_tol) || (radius_gl < radius - tolerance && max_errx_g > err_tol && max_erry_g > err_tol)
while max_errx_g > err_tol && max_erry_g > err_tol
    if radius_gl < radius
        radius_gl = radius_gl + update_steps;
    else
        radius_gl = radius_gl - update_steps;
    end
    velocity_gl = sqrt(µ/radius_gl)
    x0_gl = [radius_gl, 0, 0, 0, velocity_gl, 0]
    t0_gl = 0
    ti_gl = 2*pi*radius_gl^1.5/µ^0.5  # Should result in 1 orbit
    num_steps_gl = 1000
    xs_gl = trapezoidal(f, t0_gl, ti_gl, x0_gl, num_steps_gl);
    
    err_x_g = abs.(xs[:,1] - xs_gl[:,1])
    err_y_g = abs.(xs[:,2] - xs_gl[:,2])
    max_errx_g = maximum(err_x_g)
    max_erry_g = maximum(err_y_g)
#     @printf("Radius (m) = %.4f\n", radius_gl)
    @printf("Radius (m) = %.3f, Error x = %.3f, Error y = %.3f\n", radius_gl, max_errx_g, max_erry_g)
end
```

```julia
plot()
plot!(xs_gl[:,1], xs_gl[:,2], label="predicted guess", color="red", linewidth=5)
plot!(xs[:,1], xs[:,2], label="actual", color="cyan")
```

### comparing values of the updated predicted satellite orbit with actual orbit values

```julia
plot()
err_x_g = abs.(xs[:,1] - xs_gl[:,1])
err_y_g = abs.(xs[:,2] - xs_gl[:,2])
max_errx_g = maximum(err_x_g)
max_erry_g = maximum(err_y_g)
@printf("Updated maximum error in x (meters): %f\n", max_errx_g)
@printf("Updated maximum error in y (meters): %f\n", max_erry_g)
plot(err_x_g, err_y_g)
```
